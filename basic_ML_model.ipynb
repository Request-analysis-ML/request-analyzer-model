```python
import random

def create_sequence(arr, randVal):
    if (randVal < 1):
        val = random.randint(0,9)
        for i in range (10):
            arr.append(val)
    else:    
        for i in range (0, 10):
            arr.append(i)
    return arr    
```

```python
sequence_matrix = []
for i in range (1000):
    randVal = random.randint(0,9)
    arr = []
    sequence_matrix.append(create_sequence(arr, randVal))

sequence_df = pd.DataFrame(sequence_matrix, columns=['req0', 'req1', 'req2', 'req3', 'req4', 'req5', 'req6', 'req7', 'req8', 'req9'])
sequence_df
```
