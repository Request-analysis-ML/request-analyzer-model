```python
import pandas as pd
```

```python
df = pd.read_csv('csv_files/calculations.csv')
df
```

```python
df_spammers_scrapers = df.loc[(df['user'] == ' user-1003') | (df['user'] == ' user-1004') | (df['user'] == ' user-1005')]
```

```python
df_spammers_scrapers
```

```python
df_normal = pd.concat([df, df_spammers_scrapers]).drop_duplicates(keep = False)
```

```python
df_normal
```

```python
df_train_spammers = df_spammers_scrapers.sample(frac = 0.002)
df_test_spammers = pd.concat([df_train_spammers, df_spammers_scrapers]).drop_duplicates(keep = False)
```

```python
df_train_normal = df_normal.sample(frac = 0.8)
df_test_normal = pd.concat([df_train_normal, df_normal]).drop_duplicates(keep = False)
```

```python
df_train = pd.concat([df_train_spammers, df_train_normal])
df_test = pd.concat([df_test_spammers, df_test_normal])

df_train.to_csv('csv_files/train.csv', index=False) 
df_test.to_csv('csv_files/test.csv', index=False)
```
