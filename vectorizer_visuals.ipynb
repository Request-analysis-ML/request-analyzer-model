```python
import api_files.ml_library as calc
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer, HashingVectorizer
```

```python
"""Code for fetching the data"""

#Fetch data from csv file
data = calc.read_csv_file('csv_files/requests.csv')

#Extracts a list of all unique users in dataframe
users = calc.extract_users(data)

#Create a datachunk for a spammer
df_chunks = calc.split_user_df(data, users[0]) 
df = df_chunks[0]
```

```python
from sklearn.feature_extraction.text import HashingVectorizer
corpus = [
     'This is the first document.',
     'This document is the second document.',
     'And this is the third one.',
     'Is this the first document?',
]
vectorizer = HashingVectorizer(n_features=2**4, norm=None, alternate_sign=False)
X = vectorizer.fit_transform(corpus)
df = pd.DataFrame(X.T.todense())
df
```

```python
#vectorizers
count_vect = CountVectorizer()
Tfidf_vect = TfidfVectorizer()
hash_vect = HashingVectorizer(n_features=100, norm=None, alternate_sign=False)

#cleaning data and fit the vectorizers
clean_data = calc.clean_reqlogs(data)
count_vect = count_vect.fit(clean_data['request_logs'])
Tfidf_vect = Tfidf_vect.fit(clean_data['request_logs'])

#clean data chunk that will be transformed by the vectorizers
cleaned = calc.clean_reqlogs(df)
result_count = count_vect.transform(cleaned['request_logs'])
result_tfidf = Tfidf_vect.transform(cleaned['request_logs'])
result_hash = hash_vect.fit_transform(cleaned['request_logs'])
```

```python
cleaned['request_logs']
```

```python
#HashVectorizer for user-1003 (spammer)
df = pd.DataFrame(result_hash.T.todense(), columns=["hash"])
df.sort_values(by=["hash"],ascending=False)
```

```python
#CountVectorizer for user-1003 (spammer)
df = pd.DataFrame(result_count.T.todense(), index=count_vect.get_feature_names_out(), columns=["count"])
df.sort_values(by=["count"],ascending=False)
```

```python
#TfidfVectorizer for user-1003 (spammer)
df = pd.DataFrame(result_tfidf.T.todense(), index=Tfidf_vect.get_feature_names_out(), columns=["tfidf"])
df.sort_values(by=["tfidf"],ascending=False)
```
